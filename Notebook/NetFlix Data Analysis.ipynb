```python
df["director"] = df["director"].fillna("Unknown")
```

```python
df["duration"] = df["duration"].fillna("Not Available")
```

```python
df["cast"] = df["cast"].fillna("Not Available")
df["country"] = df["country"].fillna("Unknown")
```

```python
df["date_added"] = df["date_added"].str.strip()
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
df["date_added"].isnull().sum()
```

```python
df = df.dropna(subset=["date_added"])
```

```python
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
```

```python
df
```

```python
df["duration_int"] = df["duration"].str.extract("(\d+)")
df["duration_int"] = pd.to_numeric(df["duration_int"])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["listed_in"] = df["listed_in"].str.split(", ")
df_genre = df.explode("listed_in")
```

```python
df_genre["country"] = df_genre["country"].str.split(", ")
df_final = df_genre.explode("country")
```

```python
columns_to_keep = [
    "show_id",
    "type",
    "title",
    "country",
    "release_year",
    "year_added",
    "month_added",
    "rating",
    "duration_int",
    "listed_in"
]

df_clean = df_final[columns_to_keep]
```

```python
df_clean
```

```python
df_clean.columns = [
    "Show_ID",
    "Type",
    "Title",
    "Country",
    "Release_Year",
    "Year_Added",
    "Month_Added",
    "Rating",
    "Duration",
    "Genre"
]
```

```python
df_clean.to_csv("fact_table.csv", index=False)
```
